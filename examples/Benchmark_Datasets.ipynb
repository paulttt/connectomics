```python
import warnings
import os
import sys
root = os.path.dirname(os.path.abspath(""))+"/"
sys.path.append(root)
path = root + "example_data/benchmark/"
print(path)
from CellAnalysis.eval import benchmark
import matplotlib.pyplot as plt
from CellAnalysis.eval import benchmark
```

```python
figsize = (25, 9)
error_bands = True
```

```python
file_root = path + 'EM Data'
file_name = 'plots/EM_Results'
benchmark_em = benchmark(file_root, resolution=(0.6, 0.6, 0.6))
```

```python
#benchmark_em.print_results(metric='pq')
benchmark_em.print_results(metric='ap')
benchmark_em.print_results(metric='adc')
```

```python
benchmark_em.summarize(title='Electron Microscopy - Neuronal Cell Body Segmentation Results',
                       save_to_file=file_name,
                       error_bands=error_bands,
                       figsize=figsize,
                       file_type='pdf',
                       fontsize=18,
                       legend_fontsize=20,
                       ticksize=15)
```
